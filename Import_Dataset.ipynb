```
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

np.random.seed(42)

# Configurações
n_os = 1800
estados = ["RJ", "SP", "MG", "ES"]
tipos_servico = ["Manutenção", "Instalação", "Suporte", "Upgrade"]
prioridades = ["Baixa", "Média", "Alta", "Crítica"]

# Técnicos
tecnicos = []
for estado, qtd in zip(estados, [8, 8, 5, 4]):
    for i in range(qtd):
        tecnicos.append(f"Tecnico_{estado}_{i+1}")

# Função SLA por prioridade
def sla_por_prioridade(p):
    if p == "Baixa": return 72
    if p == "Média": return 48
    if p == "Alta": return 24
    return 8

# Base principal
dados = []

for i in range(n_os):

    estado = random.choice(estados)
    tipo = np.random.choice(tipos_servico, p=[0.4, 0.3, 0.2, 0.1])
    prioridade = random.choice(prioridades)
    tecnico = random.choice([t for t in tecnicos if estado in t])

    data_abertura = datetime(2025, 1, 1) + timedelta(days=random.randint(0, 364))
    sla = sla_por_prioridade(prioridade)

    tempo_real = np.random.normal(loc=sla*0.8, scale=5)
    atraso = tempo_real > sla

    data_fechamento = data_abertura + timedelta(hours=max(1, tempo_real))

    valor_servico = round(np.random.uniform(150, 1200), 2)
    custo = round(valor_servico * np.random.uniform(0.4, 0.7), 2)

    dados.append([
        i+1,
        estado,
        tipo,
        prioridade,
        tecnico,
        data_abertura,
        data_fechamento,
        sla,
        "Atrasado" if atraso else "Concluído",
        valor_servico,
        custo
    ])

colunas = [
    "ID_OS", "Estado", "Tipo_Servico", "Prioridade",
    "Tecnico", "Data_Abertura", "Data_Fechamento",
    "SLA_Horas", "Status", "Valor_Servico", "Custo_Operacional"
]

df = pd.DataFrame(dados, columns=colunas)

df.to_excel("base_conectafiber_2025.xlsx", index=False)

print("Base gerada com sucesso!")
```
